```python
from langchain.document_loaders import DirectoryLoader, PDFMinerLoader, TextLoader

# Specify the directory containing your PDF and TXT files
directory = '/content/data'

def load_documents(directory):
    """Loads PDF and TXT files from a directory using LangChain loaders."""

    loader = DirectoryLoader(
        directory,
        glob="**/*.pdf",  # Load PDF files
        loader_cls=PDFMinerLoader
    )
    pdf_documents = loader.load()

    loader = DirectoryLoader(
        directory,
        glob="**/*.txt",  # Load TXT files
        loader_cls=TextLoader
    )
    txt_documents = loader.load()

    # Combine PDF and TXT documents into a single list
    all_documents = pdf_documents + txt_documents

    return all_documents

documents = load_documents(directory)
print(f"Total number of documents loaded: {len(documents)}")
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

def split_docs(documents,chunk_size=500,chunk_overlap=20):
  text_splitter = RecursiveCharacterTextSplitter(chunk_size=chunk_size, chunk_overlap=chunk_overlap)
  docs = text_splitter.split_documents(documents)
  return docs

docs = split_docs(documents)
print(len(docs))
```

```python
import os
import pinecone
from langchain.vectorstores import Pinecone
from langchain.embeddings import SentenceTransformerEmbeddings

embeddings = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2")

# Set the Pinecone API key as an environment variable
os.environ["PINECONE_API_KEY"] = "47c3d32b-6fcc-4874-a6bd-cb392f37a9b3"
os.environ["PINECONE_ENVIRONMENT"] = "us-east-1"

index_name = "langchainchatbot"

# Now, Pinecone and LangChain will automatically pick up the API key and environment
index = Pinecone.from_documents(
    docs, embeddings, index_name=index_name
)
```

```python
def get_similiar_docs(query,k=1,score=False):
  if score:
    similar_docs = index.similarity_search_with_score(query,k=k)
  else:
    similar_docs = index.similarity_search(query,k=k)
  return similar_docs

query = "How is karnataka socio cultural life"
similar_docs = get_similiar_docs(query)
similar_docs
```
